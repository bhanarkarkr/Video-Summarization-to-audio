```python
# Step 2: Video to Audio conversion
VidClip = mp.VideoFileClip(r"C:\Users\LENOVO\Desktop\jn\vivekmodi.mp4")
VidClip.audio.write_audiofile(r"C:\Users\LENOVO\Desktop\jn\conv_vivekmodi.wav")
# D:\miniproject\testingVideo
```

```python
# Step 3: Speech recognition
reco = sr.Recognizer()
audio = sr.AudioFile(r"C:\Users\LENOVO\Desktop\jn\conv_vivekmodi.wav")
with audio as source:
  audio_file = reco.record(source)
result = reco.recognize_google(audio_file)
```

```python
# Step 4: Finally exporting the result
with open(r'C:\Users\LENOVO\Desktop\jn\SpeechText1.txt',mode ='w') as file:

   file.write(result)
   print("Text file ready!")
```
